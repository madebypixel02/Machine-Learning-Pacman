```python
from deepQLearning import QAgent
import matplotlib.pyplot as plt

qagent = QAgent(gamma=0.1, epsilon=1, lr=0.1, input_dims=24, batch_size=32, n_actions=4,
                    max_mem_size = 10000, eps_end=0.01,  eps_dec=5e-4, model_name="agent.net", hidden_dim=64)
qagent.recover_agent("agent.net")

qnetwork = qagent.Q_eval

print("Total experiences", qagent.total_episodes)
```

```python
# plot the network layers

fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(20, 10))
for i, layer in enumerate([qnetwork.fc1, qnetwork.fc2, qnetwork.fc3]):
    # print the activation layer
    axes[i].imshow(layer.weight.data.cpu().numpy(), cmap='Greens', interpolation='nearest', label=f'weights {i}')
```
